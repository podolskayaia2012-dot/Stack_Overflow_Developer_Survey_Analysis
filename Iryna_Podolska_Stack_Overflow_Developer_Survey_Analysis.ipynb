```python
#Імпортуємо потрібну бібліотеку
import pandas as pd
```

```python
#Завантажуємо файли з результатами і з переліком запитань
df= pd.read_csv('results.csv', low_memory=False)
df_schema= pd.read_csv('schema.csv', low_memory=False)
```

Завдання 1. Підрахунок загальної кількості респондентів

```python
#Визначаємо загальну кількість респондентів, які взяли участь в опитуванні
total_respondents = df['ResponseId'].nunique()
print(f"Всього респондентів: ", total_respondents)
```

Завдання 2. Аналіз повноти відповідей респондентів

```python
#Створюємо множину унікальних запитань із таблиці schema з їхнім переліком
questions = set(df_schema['qname'].unique())
```

```python
#Робимо перетин множини запитань із файлу з запитаннями і відповідей за 2025 рік. 
#Записуємо результат у комірку common_questions
common_questions = list(questions.intersection(df.columns))
```

```python
#Створюємо таблицю на основі первинної таблиці з відповідями. Залишаємо лише ті стовпчики, у яких немає пропусків
df_clean = df.dropna(subset=common_questions)
```

```python
#Виводимо результати підрахунку стовпчиків без пропусків 
print(f"Кількість респондентів, які дали відповіді на всі запитання: ", df_clean.shape[0])
```

```python
#Висновок: Жодний із респондентів не відповів на всі питання. 
```

Завдання 3. Статистичний аналіз досвіду респондентів

```python
#Подивимося, скільки респондентів не відповіли на питання щодо досвіду роботи або мають досвід менше 6 місяців
missing_exp = df['WorkExp'].isna().sum()
print(f"Пропущених значень у стовпці досвіду роботи: {missing_exp}")
```

```python
#Приберемо ці рядки з таблиці, щоб не викривлювати результати обчислення мір центральної тенденції
clean_exp = df.dropna(subset=['WorkExp'])
```

```python
#Підраховуємо окремо середнє, медіану і моду у первинній таблиці, виводимо результати через f-рядок.

mean = df ['WorkExp'].mean()
median = df ['WorkExp'].median()
mode= df ['WorkExp'].mode()

print(f"Досвід роботи респондентів у роках (таблиця з пропусками)\nСереднє значення:  {mean.round()} \nМедіана:  {median} \nМода:  {mode[0]}")
```

```python
#Підраховуємо окремо середнє, медіану і моду у таблиці без пропусків, виводимо результати через f-рядок.

mean = clean_exp ['WorkExp'].mean()
median = clean_exp ['WorkExp'].median()
mode= clean_exp ['WorkExp'].mode()

print(f"Досвід роботи респондентів у роках (таблиця без пропусків)\nСереднє значення:  {mean.round()} \nМедіана:  {median} \nМода:  {mode[0]}")
```

```python
#Значення за обома таблицями не відрізняються.
#Це свідчить про те, що респонденти з будь-яким досвідом роботи не давали відповідь на це питання однаково часто.
```

Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
#Залишимо тільки потрібні нам колонки
df10 = df [['ResponseId', 'Industry', 'ConvertedCompYearly', 'RemoteWork']]
```

```python
#Приберемо рядки, які мають пусті значення
df10_clean = df10.dropna().reset_index(drop=True)
```

```python
#Визначимо користувачів, які входять до 25% тих, хто заробляє більше решти
top_25 = df10_clean['ConvertedCompYearly'].quantile(0.75)
#Створимо маски для фільтрації даних за зарплатою та віддаленою роботою
top_25_devs = df10_clean['ConvertedCompYearly'] > top_25
remote_devs = df10_clean['RemoteWork'].str.contains("Remote")
```

```python
#Порахуємо кількість віддалених працівників у кожній сфері та упорядкуємо дані за найбільш поширеною сферою
best_paid_remote_industry = df10_clean [top_25_devs & remote_devs].groupby('Industry').size().sort_values(ascending=False)
best_paid_remote_industry
```
